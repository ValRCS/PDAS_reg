Second (THE BIGER) issue: 
Scripts izmet "StaleElementReferenceException", kad tiek mēģināts .txt failā ierakstīt, kaut ko vairāk no WEB lapas (piem., scriptā iekomentēto). 
No grāmatas "Web Scraping with Python": "An element in the DOM when the page initially loads, then repeatedly calling that element until Selenium throws a StaleElementReferenceException ; that is, the element is no longer attached to the page’s DOM"

```python
from selenium import webdriver
import time
from selenium.webdriver.remote.webelement import WebElement
from selenium.common.exceptions import StaleElementReferenceException
def waitForLoad(driver):
    elem = driver.find_element_by_tag_name("html")
    count = 0
    while True:
        count += 1
        if count > 20:
            print("Timing out after 10 seconds and returning")
            return
        time.sleep(.5)
        try:
            elem == driver.find_element_by_tag_name("html")
        except StaleElementReferenceException:
            return
```
